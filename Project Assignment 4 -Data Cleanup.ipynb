```python
def list_new(x):
    if x== 'As Expected':
        return '2'
    if x== 'Worse':
        return '1'
    if x== 'Better':
        return '3'
    if x== 'Unspecified':
        return '0'
result=map(list_new,unclean_data['Hospital Ratings'])
result=list(result)
for i in unclean_data.index:
    unclean_data.loc[i,'Hospital Ratings']=result[i]
unclean_data['Hospital Ratings']
```

```python
unclean_data['# of Adverse Events'] = unclean_data['# of Adverse Events'].astype(float)
unclean_data['# of Cases'] = unclean_data['# of Cases'].astype(float)

# New column for adverse event rate 
unclean_data['Adverse Event Rate'] = unclean_data['# of Adverse Events'] / unclean_data['# of Cases'] * 100
unclean_data['Adverse Event Rate'] = unclean_data['Adverse Event Rate'].fillna('0.0')

# New column for hospital latitude bucket
def get_lat_bucket(lat):
    if lat <= 36:
        return 'South'
    elif lat <= 39: 
        return 'Central'
    else:
        return 'North'

unclean_data['Lat Bucket'] = unclean_data['Latitude'].apply(get_lat_bucket)

# New column for hospital longitude bucket
def get_long_bucket(long):
    if long <= -123:
        return 'West' 
    elif long <= -119:
        return 'Central'
    else:
        return 'East'
    
unclean_data['Long Bucket'] = unclean_data['Longitude'].apply(get_long_bucket)
```
